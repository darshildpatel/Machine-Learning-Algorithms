## Gradient Descent

```python
import numpy as np
eta = 0.001
x = -4
def df(x):
    return 8*(np.power(x,3)) - 6*((np.power(x,2))) - 24*(x)
def f(x):
    return 2*(np.power(x,4)) - 2*(np.power(x,3)) - 12*(np.power(x,2)) + 6
def GradientDescent(eta, x, steps):
    for i in range(steps):
        print("Value of x before iteration", i+1,": ",x)
        print("Value of f'(x) before iteration ",i+1,": ",f(x))
        x = x + (-eta)*(df(x))
        print("Value of x after iteration", i+1,": ",x)
        print("Value of f'(x) after iteration",i+1,": ",f(x))
    
GradientDescent(eta, -4, 6)    
    
```
